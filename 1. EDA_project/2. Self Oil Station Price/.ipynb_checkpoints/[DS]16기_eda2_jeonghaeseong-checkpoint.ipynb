````python
# N or Y 를 뱉어내는 함수

def additions_Y_or_N(flag):
    
    if flag == "off":
        return 'N'
    elif flag == "셀프주유소":
        return 'Y'
    else:
        return 'Y'
````

## 1 - 4. 구의 주유소 리스트에서 정보 추출 - 전체

#### 반복문으로 전체 추출 

```
- 반복문 구조

    for in 구 리스트 :

        구 리스트 중 한개를 send_keys 하기

        검색결과 주유소 리스트를 만들기

        for in 검색결과주유소리스트:

            주유소 명 클릭

            이름추출
            주소추출
            브랜드추출
            ...
            ...      
```    

#### 반복문으로 추가하기 - 테스트로 몇 개만 해보기

````python
# 담을 리스트

st_name_list = []
st_addr_list = []
st_brand_list = []
st_gasoline_list = []
st_diesel_list = []
st_self_list = []
st_wash_list = []
st_charge_list = []
st_garage_list = []
st_store_list = []
st_allday_list = []


# 구 리스트 반복하기
for gu in notebook.tqdm(gu_names[:2]):
    
    # 구 리스트에서 구 하나 선택하기
    gu_select_tag = driver.find_element(By.ID,"SIGUNGU_NM0")
    gu_select_tag.send_keys(gu)
    time.sleep(1)
    
    # 구의 주유소 리스트를 추출하기
    gu_stations = driver.find_element(By.ID, "body1").find_elements(By.TAG_NAME,"tr")

    for station in gu_stations[:3]:
        
        # 구 리스트에서 주유소명을 클릭하기
        station.find_element(By.TAG_NAME, "a").click()
        time.sleep(1)
        
        # 현재 화면의 html 코드를 가져와서 뷰숲으로 작업하기
        req = driver.page_source
        soup = BeautifulSoup(req, 'html.parser')
        
        station_name = soup.select_one("#os_dtail_info > div > div > label").text
        st_name_list.append(station_name)

        station_addr = soup.select_one("#os_dtail_info > div > div > div > ul > li > #rd_addr")
        st_addr_list.append(station_addr.text)

        station_brand = soup.select_one("#os_dtail_info > div > div > div > ul > li > #poll_div_nm")
        st_brand_list.append(station_brand.text)

        gasoline_price = soup.select_one("#os_dtail_info > div > div > div > div > table > tbody > tr > td > #b027_p")
        st_gasoline_list.append(gasoline_price.text)

        diesel_price = soup.select_one("#os_dtail_info > div > div > div > div > table > tbody > tr > td > #d047_p")
        st_diesel_list.append(diesel_price.text)

        st_self_list.append(is_self_station())

        station_wash = soup.select_one("#os_dtail_info > div > div > div > div > #cwsh_yn")["src"][-7:-4] # "off" or "_01"
        st_wash_list.append(additions_Y_or_N(station_wash))

        station_charge = soup.select_one("#os_dtail_info > div > div > div > div > #lpg_yn")["src"][-7:-4] # "off" or "_01"
        st_charge_list.append(additions_Y_or_N(station_charge))

        station_garage = soup.select_one("#os_dtail_info > div > div > div > div > #maint_yn")["src"][-7:-4] # "off" or "_01"
        st_garage_list.append(additions_Y_or_N(station_garage))

        station_store = soup.select_one("#os_dtail_info > div > div > div > div > #cvs_yn")["src"][-7:-4] # "off" or "_01"
        st_store_list.append(additions_Y_or_N(station_store))

        station_allday = soup.select_one("#os_dtail_info > div > div > div > div > #sel24_yn")["src"][-7:-4] # "off" or "_01"
        st_allday_list.append(additions_Y_or_N(station_allday))
````

````python
tmp = pd.DataFrame(zip(
    st_name_list,
    st_addr_list,
    st_brand_list,
    st_gasoline_list,
    st_diesel_list,
    st_self_list,
    st_wash_list,
    st_charge_list,
    st_garage_list,
    st_store_list,
    st_allday_list,
))
tmp.columns = ["주유소명", "주소", "브랜드", "휘발유 가격", "경유 가격", 
"셀프 여부", "세차장 여부", "충전소 여부", "경정비 여부", "편의점 여부", "24시간 운영 여부"]
tmp
````

````python
# 담을 리스트

st_name_list = []
st_addr_list = []
st_brand_list = []
st_gasoline_list = []
st_diesel_list = []
st_self_list = []
st_wash_list = []
st_charge_list = []
st_garage_list = []
st_store_list = []
st_allday_list = []


# 구 리스트 반복하기
for gu in notebook.tqdm(gu_names):
    
    # 구 리스트에서 구 하나 선택하기
    gu_select_tag = driver.find_element(By.ID,"SIGUNGU_NM0")
    gu_select_tag.send_keys(gu)
    time.sleep(1)
    
    # 구의 주유소 리스트를 추출하기
    gu_stations = driver.find_element(By.ID, "body1").find_elements(By.TAG_NAME,"tr")

    for station in gu_stations:
        
        # 구 리스트에서 주유소명을 클릭하기
        station.find_element(By.TAG_NAME, "a").click()
        time.sleep(1)
        
        # 현재 화면의 html 코드를 가져와서 뷰숲으로 작업하기
        req = driver.page_source
        soup = BeautifulSoup(req, 'html.parser')
        
        station_name = soup.select_one("#os_dtail_info > div > div > label").text
        st_name_list.append(station_name)

        station_addr = soup.select_one("#os_dtail_info > div > div > div > ul > li > #rd_addr")
        st_addr_list.append(station_addr.text)

        station_brand = soup.select_one("#os_dtail_info > div > div > div > ul > li > #poll_div_nm")
        st_brand_list.append(station_brand.text)

        gasoline_price = soup.select_one("#os_dtail_info > div > div > div > div > table > tbody > tr > td > #b027_p")
        st_gasoline_list.append(gasoline_price.text)

        diesel_price = soup.select_one("#os_dtail_info > div > div > div > div > table > tbody > tr > td > #d047_p")
        st_diesel_list.append(diesel_price.text)

        st_self_list.append(is_self_station())

        station_wash = soup.select_one("#os_dtail_info > div > div > div > div > #cwsh_yn")["src"][-7:-4] # "off" or "_01"
        st_wash_list.append(additions_Y_or_N(station_wash))

        station_charge = soup.select_one("#os_dtail_info > div > div > div > div > #lpg_yn")["src"][-7:-4] # "off" or "_01"
        st_charge_list.append(additions_Y_or_N(station_charge))

        station_garage = soup.select_one("#os_dtail_info > div > div > div > div > #maint_yn")["src"][-7:-4] # "off" or "_01"
        st_garage_list.append(additions_Y_or_N(station_garage))

        station_store = soup.select_one("#os_dtail_info > div > div > div > div > #cvs_yn")["src"][-7:-4] # "off" or "_01"
        st_store_list.append(additions_Y_or_N(station_store))

        station_allday = soup.select_one("#os_dtail_info > div > div > div > div > #sel24_yn")["src"][-7:-4] # "off" or "_01"
        st_allday_list.append(additions_Y_or_N(station_allday))
````
